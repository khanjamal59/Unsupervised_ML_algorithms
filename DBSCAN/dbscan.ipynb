```python
from sklearn.cluster import DBSCAN
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.array([[1,2],[1,4],[2,4],[2,6],[2,2],[3,5],[3,9],
           [8,8],[9,10],[7,3],[20,20],[40,40],[40,42],[40,45],[41,42],[42,40]])
model=DBSCAN(eps=2,min_samples=2)
labels=model.fit_predict(X)
print(labels)
```

```python
plt.scatter(X[:,0],X[:,-1],X[:,1],c=labels)
plt.show()
```
